```python
import requests
import pandas as pd
```

```python
url='http://127.0.0.1:8000/n_chunks'
response=requests.get(url)
response.json()['chunks']
```

```python
url='http://127.0.0.1:8000/chunk/4'
response=requests.get(url)
json_text=response.text
short=json_text[1:-1]
short=short.replace('\\','')
```

```python
df=pd.read_json(short,orient='records')
df.head()
```

```python
df.info()
```

```python
url='http://127.0.0.1:8000/chunk/5'
response=requests.get(url)
json_text=response.text
short1=json_text[1:-1]
short1=short1.replace('\\','')
```

```python
df1=pd.read_json(short1,orient='records')
df1.head()
```

```python
df1.info()
```

```python
data=pd.concat([df,df1])
data.head()
```

```python
data.info()
```

```python
from sklearn.linear_model import TweedieRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X_train= data.drop(['gender','year','Claims1'],axis=1)
y_train= data['Claims1']


X_test = data.drop(['gender','Claims1','year'],axis=1)
y_test = data['Claims1']

reg=TweedieRegressor(power=1, alpha=0.5, link='log')
model=reg.fit(X_train,y_train)
predictions=model.predict(X_test)
```

```python
mean_squared=mean_squared_error(y_test,predictions)
mean_absolute=mean_absolute_error(y_test,predictions)#avg claims
print(f'The mean_squared_error is:{mean_squared}')
print(f'The mean_absolute_error is:{mean_absolute}')
true_and_predicted_values=pd.DataFrame({'True_values':y_test,'predicted_values':predictions})
print(true_and_predicted_values.describe())
```
